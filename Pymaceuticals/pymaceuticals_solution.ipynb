```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
study_data_complete.head(10)
```

```python
# Checking the number of mice.
len(study_data_complete["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse_ids = study_data_complete.loc[study_data_complete.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_mouse_ids[0]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = study_data_complete.loc[study_data_complete["Mouse ID"] == duplicate_mouse_ids[0]]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_study_data_comp = study_data_complete[study_data_complete["Mouse ID"].isin(duplicate_mouse_ids)==False]
clean_study_data_comp.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_study_data_comp["Mouse ID"].unique())
```

## Summary Statistics

```python
print(clean_study_data_comp.dtypes)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean = clean_study_data_comp.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
med = clean_study_data_comp.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = clean_study_data_comp.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std_dev = clean_study_data_comp.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sems = clean_study_data_comp.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
```

```python
# Assemble the resulting series into a single summary DataFrame.
summaryDict = {
    "Mean Tumor Volume": mean,
    "Median Tumor Volume": med,
    "Tumor Volume Variance": variance,
    "Tumor Volume Std. Dev.": std_dev,
    "Tumor Volume Std. Err.": sems
}

df = pd.DataFrame(summaryDict)
df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
df2 = clean_study_data_comp.groupby('Drug Regimen').agg(
    {'Tumor Volume (mm3)': ["mean", "median", "var", "std", "sem"]}
)
df2
```

## Quartiles, Outliers and Boxplots

```python
# Start by getting the last (greatest) timepoint for each mouse
maxTime = clean_study_data_comp.groupby("Mouse ID")["Timepoint"].max()
# convert to dataframe, reset index
maxTime = maxTime.reset_index()
maxTime.head()
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_data = maxTime.merge(clean_study_data_comp, on=["Mouse ID", "Timepoint"], how="left")
merge_data.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Put treatments into a list for for loop (and later for plot labels)
treat_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_list = []

for treatment in treat_list:
    # locate mice on each drug, get tumor volumes
    info = merge_data.loc[merge_data["Drug Regimen"] == treatment]
    
info
```

```python
for treatment in treat_list:
    # locate mice on each drug, get tumor volumes
    final_tumor_vol = merge_data.loc[merge_data["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    
final_tumor_vol
```

```python
tumor_list = []

for treatment in treat_list:
    # locate mice on each drug, get tumor volumes
    final_tumor_vol = merge_data.loc[merge_data["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    
    tumor_list.append(final_tumor_vol)
    
tumor_list
```

```python
tumor_list = []

for treatment in treat_list:
    # locate mice on each drug, get tumor volumes
    final_tumor_vol = merge_data.loc[merge_data["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    
    # add each series
    tumor_list.append(final_tumor_vol)
    
    # determine outliers, upper and lower bounds
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    if len(outliers) == 0:
        print(f"{treatment}'s potential outliers: None")
    else:
        output = f"{treatment}'s potential outliers: "
        for o in outliers:
            output += f"{o} "
        print(output)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# customize outliers
flierprops = dict(marker='o', markerfacecolor='red', markersize=16, markeredgecolor='black')

# customize median line
medianprops = dict(linestyle='-', linewidth=2.5, color='gold')

plt.boxplot(tumor_list, labels = treat_list, flierprops=flierprops, medianprops=medianprops, showmeans=True)

plt.ylabel('Final Tumor Volume (mm3)')
plt.title('Distrubution of the Tumor Volume for Each Treatment Group')
plt.show()
```

## Line and Scatter Plots

```python
capomulinDF = clean_study_data_comp.loc[clean_study_data_comp["Drug Regimen"] == "Capomulin"]
capomulinDF.head()
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
l509Data = capomulinDF.loc[capomulinDF["Mouse ID"] == "l509"]

plt.plot(l509Data["Timepoint"],
        l509Data["Tumor Volume (mm3)"],
        color="red")
plt.title("Capomulin Treatment of Mouse l509")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# group by mouse id, get mean
capomulinAvg = capomulinDF.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
capomulinAvgWeight = capomulinDF.groupby('Mouse ID')['Weight (g)'].mean()
```

```python
capomulinAvgDict = {
    "Tumor Volume (mm3)": capomulinAvg,
    "Weight (g)": capomulinAvgWeight
}

capomulinAvgDF = pd.DataFrame(capomulinAvgDict)
capomulinAvgDF.head(2)
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
plt.scatter(
    capomulinAvgDF["Weight (g)"],
    capomulinAvgDF["Tumor Volume (mm3)"],
    facecolor="red",
    edgecolor="black",
    s=50
)
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs Mouse Weight for Capomulin Regimen")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient
corr = st.pearsonr(capomulinAvgDF['Weight (g)'],capomulinAvgDF['Tumor Volume (mm3)'])[0]
print(f"The correlation between mouse weight and the average tumor volume is {corr:.2f}")
```

```python
# linear regression model for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = capomulinAvgDF["Weight (g)"]
y_values = capomulinAvgDF["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(capomulinAvgDF["Weight (g)"],capomulinAvgDF["Tumor Volume (mm3)"])

regressValues = x_values * slope + intercept
lineEQ = f"y = {slope:.2f}x + {intercept:.2f}"
```

```python
plt.scatter(
    x_values,
    y_values,
    facecolor="blue",
    edgecolor="black",
    s=50
)

plt.plot(x_values, regressValues, "r-")
plt.annotate(lineEQ, (20, 37), fontsize=16, color="blue")

plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs Mouse Weight for Capomulin Regimen")
plt.show()
```
